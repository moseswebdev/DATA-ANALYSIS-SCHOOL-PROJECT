Exploratory Data Analysis (EDA)

```python
plt.figure(figsize=(10, 6))
sns.histplot(climate_df1['#date+year'], bins=50, kde=True, color='blue')
plt.title('')
plt.xlabel('#date+year')
plt.ylabel('#indicator+value+num')
plt.show()
```

Correlation Analysis

```python
# Correlation heatmap
numeric_df = climate_df1.select_dtypes(include=[np.number])

# Step 2: Compute the correlation matrix
corr_matrix = numeric_df.corr()

# Step 3: Plot the correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, cmap='viridis', annot=True, fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap for Climate Data')
plt.show()
```
